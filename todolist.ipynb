```python

todo_list = []

while True:
    print("\n--- TO-DO LIST ---")
    print("1. View Tasks")
    print("2. Add Task")
    print("3. Complete Task")
    print("4. Exit")

    choice = input("\nChoose an option (1-4): ")

    if choice == "1":
        if len(todo_list) == 0:
            print("\nYour list is empty!")
        else:
            print("\nYour Tasks:")
            for index, task in enumerate(todo_list, start=1):
                status = "[Done]" if task["completed"] else "[ ]"
                print(f"{index}. {status} {task['description']}")

    elif choice == "2":
        desc = input("Enter a new task: ")
        new_task = {
            "description": desc,
            "completed": False
        }
        todo_list.append(new_task)
        print("Task added!")

    elif choice == "3":
        if len(todo_list) == 0:
            print("\nNo tasks to complete.")
        else:
            for index, task in enumerate(todo_list, start=1):
                status = "[Done]" if task["completed"] else "[ ]"
                print(f"{index}. {status} {task['description']}")

            task_num = int(input("Enter the task number to complete: "))
            actual_index = task_num - 1

            todo_list[actual_index]["completed"] = True
            print("Task marked as completed!")

    elif choice == "4":
        print("\nExit.")
        break

    else:
        print("\nInvalid choice. Try again.")
```
